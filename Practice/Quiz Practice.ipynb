Question 1: Use an if/elif/else block to test whether a string belongs to one of two lists. Depending on the outcome (belongs to list A, belongs to list B, doesn't belong to either list) print an appropriate message to the screen.

```python

list_A = ["apple", "banana", "cherry"]
list_B = ["carrot", "broccoli", "lettuce"]

word = "banana"

if word in list_A:
    print("belongs to list A")
elif word in list_B:
    print("belongs to list B")
else:
    print("doesn't belong to either list")
```

Question 2: You are given two lists of numbers, x_vals and y_vals, and must create a Numpy array where the components are related to those in the lists (for example, you might be asked to create a Numpy array with entries  ). Then generate a histogram of the values in the Numpy array, with labeled axes and a title. You will also be asked to change the color of the bars in the histogram.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x_values = [2, 3, 2, 5, 2, 3, 7, 8]
y_values = [5, 7, 2, 3, 6, 4, 1, 9]

x = np.array(x_values)
y = np.array(y_values)

z = x**2 - 5*np.sin(y)

plt.hist(z, color = "purple" )
plt.title("Histogram")
plt.xlabel("Z Values")
plt.ylabel("Frequency")
plt.show
```

Question 3: Generate random samples from two specified distributions and plot their histograms, side by side in a single figure. Each sub-figure will have its own labeled axes and title.

```python
# Norm/Uniform
fig, subfig_array = plt.subplots(nrows = 1, ncols = 2, figsize = (6, 3))

norm = np.random.normal(loc = 0, scale = 1, size = 1000)
subfig_array[0].hist(x = norm)
subfig_array[0].set_title("Normal Distribution \n (mean = 0, s.d = 1)")
subfig_array[0].set_xlabel("Sample Value")
subfig_array[0].set_ylabel("Frequency")

unif = np.random.uniform(low = 0, high = 1, size = 1000)
subfig_array[1].hist(x = unif)
subfig_array[1].set_title("Uniform Distribution \n over [0,1]")
subfig_array[1].set_xlabel("Sample Value")
subfig_array[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# # Norm/Chi
fig, subfig_array = plt.subplots(nrows = 1, ncols = 2, figsize = (6, 3))

norm = np.random.normal(loc = 0, scale = 1, size = 1000)
subfig_array[0].hist(x = norm)
subfig_array[0].set_title("Normal Distribution \n (mean = 0, s.d = 1)")
subfig_array[0].set_xlabel("Sample Value")
subfig_array[0].set_ylabel("Frequency")

chi = np.random.chisquare(df = 4, size = 1000)
subfig_array[1].hist(x = chi)
subfig_array[1].set_title("Chi-Square Distribution \n with 4 Degrees of Freedom")
subfig_array[1].set_xlabel("Sample Value")
subfig_array[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Question 6: Create a list of .csv file names and an additional empty list. Then in a for loop, read in each file to make a DataFrame, compute the mean of one of the variables/features/columns, and store it as an element of the originally empty list. Then print the list.
